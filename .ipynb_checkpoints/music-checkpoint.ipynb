```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
sns.set()
```

## Load And Read The Data Receive Into a Pandas DataFrame

```python
data = pd.read_csv('music.csv')
data.tail()
```

## Analyze Dataset

```python
data.describe()
```

```python
data.info()
```

```python
data.mean()
```

## Popularity Vs Tempo 

```python
x = data['Popularity'].values
y = data['tempo'].values
plt.figure() # Space to Plot
plt.plot(x, label='Popularity', color='red')
plt.plot(y, label='tempo')
plt.legend() # To Show The Labels
plt.title('Rock Music')
plt.xlabel('Popularity')
plt.ylabel('tempo')
plt.show() # Show The Plot
```

## What Are The Highest Liveness ?

```python
sns.violinplot( x1 = 'Artist Name' , y1 = 'liveness' , data = data)
```
